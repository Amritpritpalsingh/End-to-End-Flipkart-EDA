```python
def display_type(s):
    if(re.search(r'\)\s*(.*)$', s)):
        return re.search(r'\)\s*(.*)$', s).group(1)
    else:
        return np.nan
    
flip["Display_Type"] = flip["Display"].apply(display_type)
```

```python
flip["Display_Type"].isna().sum()
```

```python
def display_size(s):
    if(float(re.search(r'\(([\d.]+)', s).group(1))):
        return float(re.search(r'\(([\d.]+)', s).group(1))
    else:
        np.nan
flip["Display_Size"] = flip["Display"].apply(display_size)
    
```

```python
flip["Ram"] = flip["Memory"].str.split("|").str[0]
```

```python
flip.loc[flip["Ram"].str.contains("ROM", na=False), "Ram"] = np.nan
```

```python
mask = flip["Ram"].str.contains("MB", na=False)

flip.loc[mask, "Ram"] = (
    flip.loc[mask, "Ram"]
    .str.replace("MB RAM", "", regex=False)
    .str.strip()
    .astype(float)
    .div(1024)
    .astype(object)
)
```

```python
flip["Ram"]=flip["Ram"].str.replace("GB RAM ","")

```

```python
flip["Ram"] = flip["Ram"].str.replace(
    r'Expandable\s+Upto\s+\d+\s*GB',
   '',
    regex=True,
    case=False
)
flip["Ram"] = flip["Ram"].replace(r'^\s*$', np.nan, regex=True)
```

```python
flip["Ram"].isna().sum()
```

```python
flip["Ram"] = flip["Ram"].astype(float)
```

```python
flip["Ram"]=flip["Ram"].fillna(0)
```

```python
flip
```

```python
flip["Rom"] = flip["Memory"].str.extract(r'\|\s*(.*)')
```

```python

flip["Rom"] = (
    flip["Rom"]
    .str.replace(
        r'\|\s*Expandable\s+Upto\s+\d+\s*(GB|TB)',
        '',
        regex=True,
        case=False
    )
    .str.replace(r'\|\s*$', '', regex=True)  # remove trailing |
    .str.strip()
    .replace('', np.nan)
)
```

```python
flip["Rom"]=flip["Rom"].str.replace("MB ROM","")

```

```python
flip["Rom"] = flip["Rom"].fillna(0)
```

```python
flip["Rom"] = flip["Rom"].astype(float)
```
